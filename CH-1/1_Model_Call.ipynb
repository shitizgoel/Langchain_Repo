```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-5-mini", temperature=0, max_tokens=512)

response = llm.invoke('Hello! Whats your name?')
print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

response = llm.invoke('Hello! How are you?')
print(response.content)
```

```python
# Here, we don't need the model provider name. Less used 
# because ChatOpenAI and ChatGoogleGenerativeAI are used in backend.
from langchain.chat_models import init_chat_model

llm = init_chat_model("gpt-5-mini", temperature=0)

response = llm.invoke('Hello! Whats your name?')
print(response.content)
```
