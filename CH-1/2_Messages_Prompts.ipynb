### Messages

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-5-mini", temperature=0, max_tokens=512)
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="Please tell a fun fact about ocean.")
]

response = llm.invoke(messages)
print(response.content)
```

### Prompts

More user friendly than messages

```python
# Dynamic in nature
# More flexible

from langchain_core.prompts import ChatPromptTemplate

# Write as list of tuples
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a good person. Please reply in {tone} tone."),
    ("human", "Please tell a fun fact about {topic}.")
])
ready_prompt = prompt.invoke({"tone": "friendly", "topic": "ocean"})

response = llm.invoke(ready_prompt)
print(response.content)
```
