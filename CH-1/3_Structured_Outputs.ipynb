### PyDantic

Like School Principal, Very strict. Data Validation and Parsing

Pydantic would give validation error if the following conditions are not met:
- All variables (except Optional) must be there with the same name.
- The data types of all variables would match.

Common data types we can use:
- int, str, float, bool, date, datetime, list[str], Optional[str], Literal['value1', 'value2']

```python
from pydantic import BaseModel, Field
from typing import Optional

class person(BaseModel):
    """Information about a person."""
    name: str = Field(description="Name of the person")
    age: int = Field(description="Age of the person")
    home_country: str = Field(description="Country where the person is from")
    high_level_description: str = Field(description="High-level description of the person")
    key_contributions: str = Field(description="Key contributions of the person in his field")
    phone: Optional[str] = Field(description="Phone number of the person", default=None)
```

```python
# Without pydantic
response1 = llm.invoke("Tell me about Sundar Pichai in 5-10 sentences")
pprint(response1.content)
```

```python
# With pydantic
llm_structured = llm.with_structured_output(person)
response2 = llm_structured.invoke("Tell me about Sundar Pichai")
pprint(response2)   # Output is pydantic object (person class)
```

```python
# Convert to dictionary
pprint(response2.model_dump())
```

```python
response3 = llm_structured.invoke("Tell me about CEO of Microsoft")
pprint(response3.model_dump())
```

### Typedict

Like School Teacher. Tells python what should be the output format. But not raise error if the output is not coming correct.

```python
from typing import TypedDict 

class person_td(TypedDict):
    name: str
    age: int
    occupation: str
    country: str

llm_td = llm.with_structured_output(person_td)
response4 = llm_td.invoke("Tell me about Sundar Pichai")
pprint(response4)   # Output is a dictionary
```

| Feature                              | TypedDict      | Pydantic (BaseModel)      |
| ------------------------------------ | -------------- | ------------------------- |
| Runtime validation                   | ❌ No           | ✅ Yes                     |
| Type hints                           | ✅ Yes          | ✅ Yes                     |
| Data parsing                         | ❌ No           | ✅ Yes                     |
| Default values                       | Limited        | ✅ Yes                     |
| Field descriptions                   | ❌ No           | ✅ Yes                     |
| JSON Schema generation               | ❌ No           | ✅ Yes                     |
| Used with `with_structured_output()` | ❌ Not directly | ✅ Yes                     |
| Performance                          | Faster         | Slightly slower           |
| Main use                             | Static typing  | Validated structured data |


TypedDict tells Python (and your IDE) what keys a dictionary should contain.

Example:

from typing import TypedDict

class Person(TypedDict):
    name: str
    age: int

Now

person = {
    "name": "John",
    "age": 25
}

looks correct.

But Python doesn't actually validate it.

This also runs:

person = {
    "name": 123,
    "age": "Twenty"
}

No runtime error is raised.

The type checker (like mypy or Pyright) may complain during development, but your program will still execute.
