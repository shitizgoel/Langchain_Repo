```python
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatOpenAI(model="gpt-5-mini", temperature=0.2, max_tokens=500)

response = llm.invoke("Tell me about Narendar Modi G in 1 line")
print(response.content)
```

### First chain

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# Task 1

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Please reply in {tone} tone."),
    ("human", "Please tell a fun fact about {topic} in 4 sentences.")
])
```

```python
# Task 2

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.2)
```

```python
# Task 3

parser = StrOutputParser()
```

```python
# Chain

chain = prompt | llm | parser

response = chain.invoke({"tone": "friendly", "topic": "mountains"})
print(response)
```

```python
# 2nd way to do the same thing - RunnableSequence
from langchain_core.runnables import RunnableSequence

chain2 = RunnableSequence(prompt, llm, parser)

response = chain2.invoke({"tone": "friendly", "topic": "mountains"})
print(response)
```

### Second Chain

```python
from langchain_core.runnables import RunnableLambda

def dict_maker(text:str)-> dict:
    return {"text": text}

custom_runnable = RunnableLambda(dict_maker)
```

```python
prompt2 = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful social media assistant."),
    ("human", "Convert the following text into a Linkedin post of around 10 sentences: {text}")
])

chain3 = prompt | llm | parser | custom_runnable | prompt2 | llm | parser
response = chain3.invoke({"tone": "funny", "topic": "trees"})
print(response)
```
